```python
from pandas import read_csv

file = "BPA_con"

file = f"itr_cia_aberta_{file}_2011-2025.csv"

path = f"Q:\\financial_data_pipelines\\data\pipelines\\formulario_informacoes_trimestrais\\processed\\filter_columns\\{file}"
df = read_csv(path, encoding="utf-8")

df.tail(3)
```

```python
# Selecionando dados de uma empresa especifica
cnpj_klbn4 = "89.637.490/0001-45"

df_klbn4 = df[df["CNPJ_CIA"] == cnpj_klbn4]

# Selecionando informação atualizada
df_klbn4 = df_klbn4[df_klbn4["ORDEM_EXERC"] == "ÚLTIMO"]

# Selecionando informação
df_klbn4 = df_klbn4[df_klbn4["CD_CONTA"] == "1"]
df_klbn4.tail(10)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

# Estilo moderno
sns.set_style("whitegrid")

# Cria a figura
plt.figure(figsize=(12, 6))

# Gráfico de barras
plt.bar(
    df_klbn4['DT_REFER'],
    df_klbn4['VL_CONTA'],
    color='#4A90E2',          # azul elegante
    edgecolor='none',
    width=0.8
)

# Título e rótulos
plt.title('Ativo Total (em milhares de R$)', fontsize=16, fontweight='bold', loc='left', color='#333333')
plt.ylabel('Valor (R$ mil)', fontsize=12)
plt.xlabel('', fontsize=12)

# Formata o eixo Y com separador de milhar
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Rotaciona o eixo X
plt.xticks(rotation=60, ha='right', fontsize=10)

# Remove bordas laterais para deixar mais clean
sns.despine(left=True, bottom=True)

# Adiciona uma leve sombra nas barras (efeito de profundidade)
for bar in plt.gca().patches:
    bar.set_linewidth(0.2)
    bar.set_alpha(0.9)

# Ajusta margens
plt.tight_layout()

# Mostra o gráfico
plt.show()
```
