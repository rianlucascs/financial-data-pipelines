# **order_columns_CD_CONTA_e_DS_CONTA**

Lógica do methodo `order_columns_CD_CONTA_e_DS_CONTA` da classe `ExtractFormularioInformacoesTrimestrais`

```python
from pandas import read_csv

df = read_csv(r"Q:\financial_data_pipelines\data\pipelines\formulario_informacoes_trimestrais\processed\filter_columns\itr_cia_aberta_DRE_con_2011-2025.csv")
df
```

```python
df = df[df["ORDEM_EXERC"] == "ÚLTIMO"]
df.tail(4)
```

```python
df = df[["CD_CONTA", "DS_CONTA"]].drop_duplicates()
```

```python
df = df[["CD_CONTA", "DS_CONTA"]]

# Ordena pelo código (garante a sequência correta)
df = df.sort_values(by="CD_CONTA")

# Função para calcular o nível (baseado nos pontos)
def nivel(codigo):
    return codigo.count('.')

df["NIVEL"] = df["CD_CONTA"].apply(nivel)

# Gera o texto hierárquico
linhas = []
for _, row in df.iterrows():
    indent = "    " * row["NIVEL"]  # 4 espaços por nível
    linhas.append(f"{indent}{row['CD_CONTA']} -> {row['DS_CONTA']}")

# Salva em arquivo de texto
with open("estrutura_cascata.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(linhas))

print("Arquivo 'estrutura_cascata.txt' gerado com sucesso!")
```
