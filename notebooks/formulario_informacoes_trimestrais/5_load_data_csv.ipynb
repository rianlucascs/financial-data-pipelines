# **Leitura dos dados processados "csv"**

```python
from datetime import date
from pandas import read_csv


def load_data_itr(path, file, cnpj=None, cd_conta=None, ds_conta=None):

    path = f"{path}\\itr_cia_aberta_{file}_2011-{date.today().year}.csv"

    df = read_csv(path, encoding="utf-8")

    if cnpj != None:

        df = df[df["CNPJ_CIA"] == cnpj]
        

        if cd_conta != None:
            df = df[df["ORDEM_EXERC"] == "ÚLTIMO"]
            df = df[df["CD_CONTA"] == cd_conta]

        if ds_conta != None:
            df = df[df["ORDEM_EXERC"] == "ÚLTIMO"]
            df = df[df["DS_CONTA"] == ds_conta]

    return df


file = "DRE_con"
path = r"Q:\financial_data_pipelines\data\pipelines\formulario_informacoes_trimestrais\interim"
cnpj_klbn = "89.637.490/0001-45"
cd_conta = "3.01"
ds_conta = "Custo dos Bens e/ou Serviços Vendidos"

df = load_data_itr(path, file, cnpj_klbn, ds_conta=ds_conta) 
```
