# **Leitura dos dados processados "csv"**

```python
from datetime import date
from pandas import read_csv


def load_data_itr(path, file, cnpj=None, cd_conta=None):

    path = f"{path}\\itr_cia_aberta_{file}_2011-{date.today().year}.csv"

    df = read_csv(path, encoding="utf-8")

    if cnpj != None:
        df = df[df["CNPJ_CIA"] == cnpj]
        df = df[df["ORDEM_EXERC"] == "ÚLTIMO"]
        if cd_conta != None:
            df = df[df["CD_CONTA"] == cd_conta]
            df = df[["DENOM_CIA", "DT_REFER", "DS_CONTA", "VL_CONTA"]].reset_index(drop=True)

    return df


file = "DRE_con"
path = r"Q:\financial_data_pipelines\data\pipelines\formulario_informacoes_trimestrais\processed\filter_columns"
cnpj_klbn = "89.637.490/0001-45"
cd_conta = "3.01"

load_data_itr(path, file, cnpj_klbn, cd_conta) 
```
