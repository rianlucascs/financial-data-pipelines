**CD_CONTA**

Cada código representa uma categoria de ativo, e a hierarquia mostra contas principais e subcontas.


### 1️⃣ Níveis da hierarquia

Cada número separado por ponto indica um **nível** da conta:

* **1** → Nível 1 → **Ativo Total** (tudo que a empresa possui)
* **1.01** → Nível 2 → **Ativo Circulante** (caixa, contas a receber, estoques — recursos que viram dinheiro rápido)
* **1.01.01** → Nível 3 → **Caixa e Equivalentes de Caixa**
* **1.01.01.01** → Nível 4 → **Caixa** (dinheiro em mãos)
* **1.01.01.02** → Nível 4 → **Aplicações Financeiras** (investimentos de curto prazo)

Quanto mais pontos e números, mais **detalhada** é a conta.


### 2️⃣ Como se organiza

* **Conta pai** → soma ou engloba os valores de todas as suas **subcontas**.
* Exemplo:

  * `1.07` → Intangível (total do ativo intangível)

    * `1.07.01` → Intangíveis

      * `1.07.01.01` → Valor de Custo dos Intangíveis
      * `1.07.01.02` → Perdas por Redução ao Valor Recuperável
    * `1.07.02` → Goodwill
    * `1.07.03` → Amortização Acumulada

> Assim, o valor da conta `1.07` = soma de todos os valores das subcontas (`1.07.01 + 1.07.02 + 1.07.03`).

**ST_CONTA_FIXA**

* Assim, `ST_CONTA_FIXA` é um **indicador lógico** (S/N ou verdadeiro/falso) que informa se uma determinada linha ou conta contábil está no “elenco de contas fixas” ou no “elenco de contas não fixas”.
* **Contas fixas** seriam aquelas que seguem um formato padronizado (mesmas denominações, mesmas posições de linha) para todas as companhias ou para grande parte delas, para garantir comparabilidade.
* **Contas não fixas** são aquelas que variam entre empresas (ou entre períodos) — ou seja: linhas específicas, customizadas, ou que não fazem parte do template padrão.
