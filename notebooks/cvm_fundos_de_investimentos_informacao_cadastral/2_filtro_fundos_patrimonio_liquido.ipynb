```python
from pandas import read_csv, set_option

path = lambda file: f"Q:\\financial_data_pipelines\\data\\pipelines\\cvm_fundos_de_investimentos_informacao_cadastral\\raw\\csv\\{file}"
df_registro_fundo = read_csv(path("registro_fundo.csv"), sep=";",  encoding="latin1")
df_registro_fundo.tail(3)
```

```python
# todos os tipos de fundo: filtro maior patrimonio liquido
df_registro_fundo_f1 = df_registro_fundo[df_registro_fundo["Situacao"] == "Em Funcionamento Normal"]
df_registro_fundo_f1 = df_registro_fundo_f1[["Denominacao_Social", "Patrimonio_Liquido"]].sort_values(by="Patrimonio_Liquido").dropna().tail(10).sort_values(by="Patrimonio_Liquido", ascending=False)
df_registro_fundo_f1["Patrimonio_Liquido"] = df_registro_fundo_f1["Patrimonio_Liquido"].apply(lambda valor: f"R$ {float(valor):,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
set_option('display.max_colwidth', None)
df_registro_fundo_f1.reset_index(drop=True)
```

```python
tipos_acoes = ["FI"]

df_registro_fundo_f2 = df_registro_fundo[
    (df_registro_fundo["Situacao"] == "Em Funcionamento Normal") &
    (df_registro_fundo["Tipo_Fundo"].isin(tipos_acoes))
]

df_registro_fundo_f2
```
